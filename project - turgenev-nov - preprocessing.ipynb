```python
def len_text (word):
    word = word.lower()
    word = re.sub("[!\"#$%&\'—()*+,-»«./:;<=>?@[\\]^_`{|}~]?",'',word)
    word = word.split()
    output = len(word)
    return output
```

```python
def remove_stopwords(Word_list):
        processed_Word_list = []
        for Word in Word_list:
            if Word not in stopwords.words("russian"):
                processed_Word_list.append(Word)
        return processed_Word_list
```
