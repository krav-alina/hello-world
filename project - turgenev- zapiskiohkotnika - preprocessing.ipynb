```python
tom3 = tom3.lower()
tom3 = re.sub("[!\"#$%&\'—()*+,-»«./:;<=>?@[\\]^_`{|}~]?",'',tom3)
tom3 = " ".join([word for word in tom3.split() if word not in(stopwords.words('russian'))])
```

```python
bigram_vectorizer = CountVectorizer(ngram_range=(2,4),
                                    token_pattern=r'[А-Яа-я]+', min_df=1)
```

```python
analyze = bigram_vectorizer.build_analyzer()
```

```python
z = analyze(tom3)
```
